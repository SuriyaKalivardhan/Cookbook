```python
auth = OpenApiAnonymousAuthDetails()

# Initialize agent OpenApi tool using the read in OpenAPI spec
openapi_tool = OpenApiTool(
    name="EmployeeService",
    spec= json.loads(employee_service),
    description="Employees service that can create, update, get, list and delete records",
    auth=auth,
    default_parameters=["format"],
)

openapi_tool.add_definition(
    name="WeatherService",
    spec=json.loads(weather_service),
    description="Retrieves the weather",
    auth=auth,
)

agent = project_client.agents.create_agent(
    model="gpt-4o-mini",
    name="employeweatherService",
    instructions="You are a helpful assistant that uses the employee service tool and weather service to serve the user questions ",
    tools=openapi_tool.definitions
)
```

```python
thread = project_client.agents.create_thread()
```
